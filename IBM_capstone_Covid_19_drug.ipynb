```python
#selecting the relevant columns

df_short=df[['TrialID','Public title', 'Scientific title','Primary sponsor','Phase','Countries','Condition']]

df_short.tail(2)
#df_short.shape                
```

### Identifying missing values , only process cells with assigned countries.

```python
df_short=df_short.dropna(subset=['Countries']) #drop field with NAN
#df_short.shape
```

```python
if 'df_short_1' in globals():
    del df_short_1
```

### Separate rows with countries that have multiple countries. Each country should have its own field.

```python
#iterate among rows to identify fields with multiple countries

    
for row_index, row in df_short.iterrows():
    country=str(row['Countries'])
    #print(country) 
    country1=country.split(';')  #splitting multiple countries fields
   # print('\nrow number:', row_index, country)

    if (len(country1)>=2) and (country1[0]!= country1[1]):
        #print(country1, country1[0],country1[1])
        for i in range(len(country1)):
            temp=df_short.loc[row_index].copy()
            temp['Countries']=country1[i]
            df_short=df_short.append(temp, ignore_index=True)
    elif (len(country1)>=2) and (country1[0]==country1[1]):
            temp=df_short.loc[row_index] 
            temp['Countries']=country1[0]
            df_short=df_short.append(temp, ignore_index=True)
            
   
```

```python
#df_short_2=df_short_1 
df_short.shape
```

```python
##delete countries with two or more listed countries 
for row_index, row in df_short.iterrows():
    countrya=str(row['Countries'])
    countryb=countrya.split(';')
    #print (countryb)
    #print(len(countryb))
    if (len(countryb)>=2):
       # print(row_index,row['Countries'])
        df_short.drop(row_index, axis=0, inplace=True)
```

```python
#reset indexes to countries 
df_short.set_index('Countries', inplace=True)
df_short.head()
```
